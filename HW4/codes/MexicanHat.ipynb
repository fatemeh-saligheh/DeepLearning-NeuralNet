```python
import numpy as np
import matplotlib.pyplot as plt
x = [ 0.27, 0.35, 0.44, 0.58, 0.66, 0.77, 0.4, 0.32, 0.20, 0.15, 0.08 ]
x = np.array(x)
```

```python
a = np.copy(x)
t_max = 4
c1 = 0.6
c2 = -0.1
R1 = 0
R2 = 20
for t in range(t_max):
    a_old = np.copy(a)
    for i in range(len(a)):
        sum1 = 0
        for j in range(1,R1+1):
            if(i-j>0):
                sum1 = sum1 + a_old[i-j]
            if(i+j<len(a)):
                sum1 = sum1 + a_old[i+j]
        sum2 = 0
        for k in range(R1+1,R2+1):
            if k > len(a) :
                break
            if(i-k>0):
                sum2 = sum2 + a_old[i-k]
            if(i+k<len(a)):
                sum2 = sum2 + a_old[i+k]
#         print('sum1:',sum1)
#         print('sum2:',sum2)
        asum = c1*a_old[i] + c1*sum1 + c2*sum2
#         print(asum)
        if(asum <= 0):
            a[i] = 0
        elif(asum>=2):
            a[i] = 2
        else:
            a[i] = asum
    plt.plot(a)
    print(a)
plt.show()
```
