```python
import numpy as np
import array
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def genSet(n1,n2):
    X1dataA=1+0.5*np.random.normal(0,1,n1)
    X2dataA=1+0.5*np.random.normal(0,1,n1)
    labels = []
    for i in range(n1):
        labels.append(1)
    X1dataB=-1+0.5*np.random.normal(0,1,n2)
    X2dataB=-1+0.5*np.random.normal(0,1,n2)
    for i in range(n2):
        labels.append(-1)
    return X1dataA,X2dataA,X1dataB,X2dataB,labels
```

```python
class Perceptron():
    def __init__(self,inputLength,lr=0.01):
        self.lr = lr
        self.weights = [0 for i in range(inputLength+1)]

    def train(self,data,labels):
        flag = True
        i = 0
        while flag and (i<1000):
            cost = 0
            i+=1
            for row,t in zip(data,labels):
                net = np.dot(row,self.weights[1:])+1*self.weights[0]
                if net > 0:
                    h = 1
                else:
                    h = -1           
                self.weights[1:] += self.lr * (t) * row
                self.weights[0] += self.lr * (t)
#                 print(t,h)
                cost = cost +abs(t-h)
            
            if(cost == 0):
                print("round =",i," cost=",cost)
                flag = False

        return self.weights
```

```python
class Adaline():
    def __init__(self,inputLength,lr=0.2):
        self.lr = lr
        self.weights = [0 for i in range(inputLength+1)]

    def train(self, data, labels):
        flag = True
        i = 0
        while flag and (i<1000):
            cost = 0
            i+=1
            for row, t in zip(data, labels):
                net = np.dot(row, self.weights[1:]) + 1*self.weights[0]
                self.weights[1:] += self.lr * (t - net) * row
                self.weights[0] += self.lr * (t - net)
                
                if net > 0:
                    h = 1
                else:
                    h = -1
                cost = cost +abs(t-h)
            if(cost < 0.4):
                print("round =",i," cost=",cost)
                flag = False   
        return self.weights
```

```python
X1dataA,X2dataA,X1dataB,X2dataB,labels = genSet(100,100)
training_inputs = []
for i in range(len(X1dataA)):
    training_inputs.append(np.array([X1dataA[i],X2dataA[i]]))
for i in range(len(X1dataB)):
    training_inputs.append(np.array([X1dataB[i],X2dataB[i]]))
perceptron = Perceptron(2,lr=0.001)
weights = perceptron.train(training_inputs, labels)
a= -1*weights[1]/weights[2]
b= -1*weights[0]/weights[2]
x = np.linspace(-2,2,4)
y=a*x+b
plt.plot(x,y)
plt.scatter(X1dataA,X2dataA) 
plt.scatter(X1dataB,X2dataB) 
plt.show()
```
